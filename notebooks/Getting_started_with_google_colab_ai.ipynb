```python
import io
import pandas as pd
df = pd.read_csv('ab_test_click_data (1).csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.head()
df.groupby("group")
```

```python
palatte = {0 :"violet" , 1 :"black"}
```

```python
plt.figure(figsize=(10,6))
```

```python
ax=sns.countplot(x="group" ,hue ="click" , data=df ,palette=palatte )
plt.title("click distribution A/B testing")
plt.xlabel("group")
plt.ylabel("count")
plt.legend(title ="click", labels=["yes" , "no"])
```
